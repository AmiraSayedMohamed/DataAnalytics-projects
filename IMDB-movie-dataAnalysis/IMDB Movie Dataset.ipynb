# Steps for Data Analysis:


---------------------------------
  ##  1.Understand the Dataset:
-------------------------

        - Review each column in the dataset to understand its meaning.
        - Identify the type of data in each column (e.g., numerical, categorical, text).
   ----------------     
 ## 2.Data Cleaning:
 ------------------

    - Handle missing values (e.g., Revenue and Metascore have NaN values).
    - Correct data types if needed (e.g., ensure 'Year' is an integer, 'Revenue' is a float).
    - Remove any duplicates if present.
    
---------------------------------
 ## 3.Exploratory Data Analysis (EDA):
--------------------------------
     
   ### Descripive Statistics:
    - Summarize numerical data (mean, median, mode, etc.).
    - Count unique values in categorical columns like Genre, Director, and Actors.
   ### Correlations:
    - Analyze the correlation between numerical variables (e.g., Rating vs. Revenue).
   ### Outliers:
    - Identify and investigate outliers in numerical columns like Revenue and Votes.
    
--------------------------------------------------
 ## 4.Formulate Questions:
 -----------------------------
 
   ### General Questions:
    - What are the most common genres?
    - Which directors and actors are most frequent in the top-ranked movies?
    - What is the average runtime of movies? 
   ### Performance-Related Questions:
    - How does movie rating correlate with revenue?
    - Is there a relationship between runtime and rating?
    - Which years produced the highest-grossing movies?
   ### Trend Analysis:
    - How have movie ratings changed over the years?
    - What genres have the highest average revenue?
    
-----------------------
   ## 5.Extract Insights:
-----------------------

    - Determine the top genres and their performance metrics (e.g., average rating and revenue).
    - Identify directors who consistently produce high-revenue movies.
    - Analyze the effect of runtime on a movieâ€™s rating and revenue.
    - Evaluate the most profitable years and the trends in movie production.

----------------------------------------------------------------------------------
  ## 6. Visualization (Graphs to Appear in Dashboard):
-----------------------------------------------------

   ### Histogram:
    - Distribution of movie ratings.
    - Distribution of revenues.
    - Distribution of runtimes.
   ### Bar Chart:
    - Count of movies by genre.
    - Average revenue per genre.
    - Top directors based on average rating and revenue.
   ### Scatter Plot:
    - Rating vs. Revenue.
    - Runtime vs. Rating.
   ### Line Plot:
    - Trends of movie ratings over the years.
    - Trends of average revenue over the years.
   ### Box Plot:
    - Distribution of ratings and revenues by genre.
   ### Heatmap:
    - Correlation matrix of numerical variables (Rating, Votes, Revenue, Metascore).
    
  
  ## 7.Dashboard Layout Using Subplots:

   ### Top Row:
        Left: Bar chart showing the count of movies by genre.
        Right: Scatter plot of Rating vs. Revenue.
   ### Middle Row:
        Left: Histogram of movie ratings.
        Right: Line plot showing trends of movie ratings over the years.
   ### Bottom Row:
    Left: Box plot showing the distribution of revenue by genre.
    Right: Heatmap of the correlation matrix.

```python
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

# Assuming 'data' is your DataFrame and 'counter_data' is your Counter object
counter_data = Counter({
        'Drama': 474,
        'Action': 289,
        'Comedy': 268,
        'Adventure': 249,
        'Thriller': 183,
        'Crime': 141,
        'Romance': 131,
        'Sci-Fi': 114,
        'Horror': 108,
        'Mystery': 99,
        'Fantasy': 98,
        'Biography': 71,
        'Family': 50,
        'Animation': 46,
        'History': 28,
        'Music': 16,
        'Sport': 16,
        'War': 12,
        'Western': 6,
        'Musical': 5
})

# Extract genres and frequencies
genres = list(counter_data.keys())
frequencies = list(counter_data.values())

# Create a figure with subplots
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(18, 12))

# Bar chart for genre distribution
axes[0, 0].bar(genres, frequencies, color='blue')
axes[0, 0].set_xlabel('Genre')
axes[0, 0].set_ylabel('Frequency')
axes[0, 0].set_title('Genre Distribution')
axes[0, 0].tick_params(axis='x', rotation=45)

# Count plot for Votes by Year (corrected)
sns.countplot(x='Year', hue='Year', palette='bright', data=data, ax=axes[0, 1])
axes[0, 1].set_title('Votes By Year')

# Scatter plot for Rating vs. Revenue
sns.scatterplot(x='Rating', y='Revenue (Millions)', data=data, ax=axes[0, 2])
axes[0, 2].set_title('Rating vs. Revenue')

# Histogram of Movie Ratings
sns.histplot(x='Rating', data=data, bins=20, ax=axes[1, 0])
axes[1, 0].set_title('Histogram of Movie Ratings')

# Line plot for Movie Ratings Over Time
sns.lineplot(x='Year', y='Rating', data=data, ax=axes[1, 1])
axes[1, 1].set_title('Movie Ratings Over Time')

# Bar plot for Votes By Year (without legend)
sns.barplot(x='Year', y='Votes', data=data, palette='bright', dodge=False, ax=axes[1, 2])
axes[1, 2].set_title("Votes By Year")
axes[1, 2].legend([],[], frameon=False)  # Hide the legend

# Histogram for Revenue Distribution
axes[2, 0].hist(data['Revenue (Millions)'], bins=5, edgecolor='black')
axes[2, 0].set_xlabel('Revenue (Millions)')
axes[2, 0].set_ylabel('Count')
axes[2, 0].set_title('Revenue Distribution')

# Line plot for Average Movie Ratings Over the Years
rating = data.groupby('Year')['Rating'].mean()
axes[2, 1].plot(rating.index, rating.values, marker='o', color='b', linestyle='-')
axes[2, 1].set_title('Average Movie Ratings Over the Years')
axes[2, 1].set_xlabel('Year')
axes[2, 1].set_ylabel('Average Rating')
axes[2, 1].grid(True)

# Hide the empty subplot (2, 2)
axes[2, 2].axis('off')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the combined plot
plt.show()
```
